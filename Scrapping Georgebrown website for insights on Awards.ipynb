```python
import requests
from bs4 import BeautifulSoup
import openpyxl
from openpyxl import Workbook
import os


url = "https://www.georgebrown.ca/apply/financial-aid/awards-and-scholarships/types-of-awards/external"
response = requests.get(url)
html = response.content


bsobj = BeautifulSoup(html, 'html.parser')

# Create a new Workbook object
workbook = Workbook()

# Get the active worksheet
worksheet = workbook.active

# Write headers to the Excel file
worksheet.cell(row=1, column=1, value="Heading")
worksheet.cell(row=1, column=2, value="Paragraph")

# Initialize a row counter
row_counter = 2

# Iterate over the headings and paragraphs
for heading in bsobj.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6']):
    paragraph = heading.find_next_sibling('p')
    if paragraph:
        # Write the heading and paragraph to the Excel file
        worksheet.cell(row=row_counter, column=1, value=heading.get_text(strip=True))
        worksheet.cell(row=row_counter, column=2, value=paragraph.get_text(strip=True))
        row_counter += 1


desktop_path = os.path.join(os.path.expanduser('~'), 'Desktop')

# Save the Excel file on the desktop
output_file = os.path.join(desktop_path, "headings_paragraphs.xlsx")
workbook.save(output_file)

print(f"Excel file saved on the desktop: {output_file}")
```
